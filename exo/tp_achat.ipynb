```python
if 'Sexe' in df.columns and 'Fréquence_Achats' in df.columns:
    depenses_moyennes = df.groupby('Sexe')['Fréquence_Achats'].mean()
    print(depenses_moyennes)
else:
    print("Les colonnes 'Sexe' ou 'Fréquence_Achats' n'existent pas.")
```

```python


if 'Catégories_Produits' in df.columns and 'Montant_Achats' in df.columns:
    depenses_par_categorie = df.groupby('Catégories_Produits')['Montant_Achats'].sum()

    depenses_par_categorie.plot(kind='bar', color='skyblue')

    plt.title('Dépenses par catégorie de produits')
    plt.xlabel('Catégorie de produits')
    plt.ylabel('Dépenses totales')

    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

else:
    print("Les colonnes 'Catégories_Produits' ou 'Montant_Achats' n'existent pas.")
```

```python
if 'Sexe' in df.columns:
    sexe_counts = df['Sexe'].value_counts()

    plt.pie(sexe_counts, labels=sexe_counts.index, autopct='%1.1f%%', colors=['skyblue', 'salmon'])

    plt.title('Proportions d\'hommes et de femmes')

    plt.show()
else:
    print("La colonne 'Sexe' n'existe pas.")
```

```python
data = {
    'CatÃ©gories_Produits': [100, 110, 120, 130, 140],
    'Age': [22, 25, 30, 28, 35]
}

df = pd.DataFrame(data)

print("Colonnes avant correction :", df.columns)

df.columns = df.columns.str.replace('CatÃ©gories_Produits', 'Catégories_Produits')

print("Colonnes après correction :", df.columns)

correlation_matrix = df[['Catégories_Produits', 'Age']].corr()

print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Matrice de Corrélation")
plt.show()
```
