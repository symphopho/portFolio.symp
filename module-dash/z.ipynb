```python
import pandas as pd
```

```python

df = pd.read_csv('gapminder_edited.csv')
df
```

```python
#  ON ne garde que les valeurs de l'annÃ©e 2007
df = df[df['year'] == 2007]
df
```

```python
df = df.sort_values(by='pop', ascending=False)
```

```python
import matplotlib.pyplot as plt
# histogram of pop by country

fig = plt.figure(figsize=(10, 6))
plt.bar(df['country'], df['pop'])
# rotation of country names
plt.xticks(rotation=90)
fig.show()
```
